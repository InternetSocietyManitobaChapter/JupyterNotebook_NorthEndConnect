#### Split the TEST DATE/UTC HOUR column into two columns.One holding TEST DATE and the other UTC HOUR

```python
northEndConnectData_TimeSplit = northEndConnectData
testDate = []
utcHour = []

for value in northEndConnectData['TEST DATE / UTC HOUR']:
    spaceIndex = value.index(" ")
    dateData = value[:spaceIndex]
    timeData = value[spaceIndex:]
    testDate.append(dateData)
    utcHour.append(timeData)
    
northEndConnectData_TimeSplit["TEST DATE"] = testDate
northEndConnectData_TimeSplit["TEST TIME"] = utcHour
```

```python
northEndConnectData_TimeSplit
```

```python
northEndConnect_SplitDate = northEndConnectData_TimeSplit

year = []
month = []
day = []

for value in northEndConnect_SplitDate["TEST DATE"]:
    first_delimiter = value.index("/")
    second_delimiter = value.index("/", first_delimiter+1)
    monthOfDate = value[:first_delimiter]
    dayOfDate = value[first_delimiter+1:second_delimiter]
    yearOfDate = value[-4:]
    
    day.append(dayOfDate)
    month.append(monthOfDate)
    year.append(yearOfDate)
    
#northEndConnect_SplitDate["YEAR"] = year
#northEndConnect_SplitDate["MONTH"] = month
#northEndConnect_SplitDate["DAY"] = day
northEndConnect_SplitDate.insert(2, "YEAR", year, True)
northEndConnect_SplitDate.insert(3, "MONTH", month, True)
northEndConnect_SplitDate.insert(4, "DAY", day, True)
```

```python
northEndConnect_SplitDate
```

```python
# Having split the "TEST DATE / UTC TIME" column into the YEAR, MONTH, DAY, and TIME columns, we can
# Drop the "TEST DATE / UTC TIME" column.

northEndConnect_SplitDate_dropTESTDATE_UTCHOUR = northEndConnect_SplitDate.drop(columns = ['TEST DATE / UTC HOUR'])
```

### Categorize the Internet Service based on Upload and Download speeds, Jitter, Latency
### and Packet Loss

```python
WinnipegArea_InternetCategory = WinnipegAreaData_from2019_AreaCode


InternetCategory = []

for value in range(len(WinnipegArea_InternetCategory)):
    if((WinnipegArea_InternetCategory.loc[value, 'DOWNLOAD SPEED'] >= 50) 
    and (WinnipegArea_InternetCategory.loc[value, 'UPLOAD SPEED'] >= 10) 
    and (WinnipegArea_InternetCategory.loc[value, 'JITTER'] <= 5) 
    and (WinnipegArea_InternetCategory.loc[value, 'LATENCY'] <= 50) 
    and (WinnipegArea_InternetCategory.loc[value, 'PACKET LOSS'] <= 0.25)):
        
        internetStatus = "High-Speed Internet"
        
    else:
        
        internetStatus = "Low-Speed Internet"
        
    InternetCategory.append(internetStatus) 
    
WinnipegArea_InternetCategory["INTERNET_CATEGORY"] = InternetCategory    
```
